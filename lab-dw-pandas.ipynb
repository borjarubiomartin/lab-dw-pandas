### Data Description

- Customer - Customer ID

- ST - State where customers live

- Gender - Gender of the customer

- Education - Background education of customers 

- Customer Lifetime Value - Customer lifetime value(CLV) is the total revenue the client will derive from their entire relationship with a customer. In other words, is the predicted or calculated value of a customer over their entire duration as a policyholder with the insurance company. It is an estimation of the net profit that the insurance company expects to generate from a customer throughout their relationship with the company. Customer Lifetime Value takes into account factors such as the duration of the customer's policy, premium payments, claim history, renewal likelihood, and potential additional services or products the customer may purchase. It helps insurers assess the long-term profitability and value associated with retaining a particular customer.

- Income - Customers income

- Monthly Premium Auto - Amount of money the customer pays on a monthly basis as a premium for their auto insurance coverage. It represents the recurring cost that the insured person must pay to maintain their insurance policy and receive coverage for potential damages, accidents, or other covered events related to their vehicle.

- Number of Open Complaints - Number of complaints the customer opened

- Policy Type - There are three type of policies in car insurance (Corporate Auto, Personal Auto, and Special Auto)

- Vehicle Class - Type of vehicle classes that customers have Two-Door Car, Four-Door Car SUV, Luxury SUV, Sports Car, and Luxury Car

- Total Claim Amount - the sum of all claims made by the customer. It represents the total monetary value of all approved claims for incidents such as accidents, theft, vandalism, or other covered events.

## Challenge 1: Understanding the data

In this challenge, you will use pandas to explore a given dataset. Your task is to gain a deep understanding of the data by analyzing its characteristics, dimensions, and statistical properties.

- Identify the dimensions of the dataset by determining the number of rows and columns it contains.
- Determine the data types of each column and evaluate whether they are appropriate for the nature of the variable. You should also provide suggestions for fixing any incorrect data types.
- Identify the number of unique values for each column and determine which columns appear to be categorical. You should also describe the unique values of each categorical column and the range of values for numerical columns, and give your insights.
- Compute summary statistics such as mean, median, mode, standard deviation, and quartiles to understand the central tendency and distribution of the data for numerical columns. You should also provide your conclusions based on these summary statistics.
- Compute summary statistics for categorical columns and providing your conclusions based on these statistics.

```python
# Your code here
import pandas as pd
import numpy as np

# Load the dataset
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df = pd.read_csv(url)

print("=== DATASET DIMENSIONS ===")
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
print(f"Dataset shape: {df.shape}")

print("\n=== DATA TYPES ===")
print(df.dtypes)
print("\nData type suggestions:")
print("- Customer: Should be string/object (categorical identifier)")
print("- ST: Should be string/object (categorical - state codes)")
print("- Gender: Should be string/object (categorical)")
print("- Education: Should be string/object (categorical)")
print("- Customer Lifetime Value: Should be numeric (float)")
print("- Income: Should be numeric (int/float)")
print("- Monthly Premium Auto: Should be numeric (float)")
print("- Number of Open Complaints: Should be numeric (int)")
print("- Policy Type: Should be string/object (categorical)")
print("- Vehicle Class: Should be string/object (categorical)")
print("- Total Claim Amount: Should be numeric (float)")

print("\n=== UNIQUE VALUES AND CATEGORICAL ANALYSIS ===")
for column in df.columns:
    unique_count = df[column].nunique()
    print(f"\n{column}:")
    print(f"  Unique values: {unique_count}")
    
    # Determine if categorical (typically if unique values < 20 or if it's clearly categorical)
    if df[column].dtype == 'object' or unique_count < 20:
        print(f"  Type: Categorical")
        print(f"  Unique values: {list(df[column].unique())}")
    else:
        print(f"  Type: Numerical")
        print(f"  Range: {df[column].min()} to {df[column].max()}")

print("\n=== NUMERICAL COLUMNS SUMMARY STATISTICS ===")
numerical_columns = df.select_dtypes(include=[np.number]).columns
print(df[numerical_columns].describe())

print("\n=== INSIGHTS ON NUMERICAL DATA ===")
for col in numerical_columns:
    mean_val = df[col].mean()
    median_val = df[col].median()
    std_val = df[col].std()
    
    print(f"\n{col}:")
    print(f"  Mean: {mean_val:.2f}")
    print(f"  Median: {median_val:.2f}")
    print(f"  Standard Deviation: {std_val:.2f}")
    
    if mean_val > median_val:
        print(f"  Distribution: Right-skewed (mean > median)")
    elif mean_val < median_val:
        print(f"  Distribution: Left-skewed (mean < median)")
    else:
        print(f"  Distribution: Approximately symmetric")

print("\n=== CATEGORICAL COLUMNS SUMMARY STATISTICS ===")
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    print(f"\n{col}:")
    print(f"  Most frequent value: {df[col].mode()[0]}")
    print(f"  Frequency of most common: {df[col].value_counts().iloc[0]}")
    print(f"  Value counts:")
    print(df[col].value_counts())

print("\n=== MISSING VALUES ===")
print(df.isnull().sum())

print("\n=== GENERAL INSIGHTS ===")
print("- The dataset contains insurance customer data with both demographic and policy information")
print("- No missing values detected")
print("- Mix of categorical and numerical variables")
print("- Customer Lifetime Value and Total Claim Amount show high variability")
print("- Gender distribution appears balanced")
print("- Multiple policy types and vehicle classes available")
```

The marketing team wants to know the top 5 less common customer locations. Create a pandas Series object that contains the customer locations and their frequencies, and then retrieve the top 5 less common locations in ascending order.

### Exercise 2

*Hint:*

- *Futhermore, there is a method that returns the index of the maximum value in a column or row.*


```python
# Your code here
# Create a pandas Series with policy types and their total number of policies sold
policy_counts = df['Policy Type'].value_counts()

print("Total number of policies sold for each policy type:")
print(policy_counts)

# Get the policy type with the highest number of policies sold
highest_policy_type = policy_counts.idxmax()
highest_policy_count = policy_counts.max()

print(f"\nPolicy type with the highest number of policies sold:")
print(f"Policy Type: {highest_policy_type}")
print(f"Number of policies: {highest_policy_count}")
```

### Exercise 3

The sales team wants to know if customers with Personal Auto have a lower income than those with Corporate Auto. How does the average income compare between the two policy types?

- Use *loc* to create two dataframes: one containing only Personal Auto policies and one containing only Corporate Auto policies.
- Calculate the average income for each policy.
- Print the results.

```python
# Your code here
# Use loc to create two dataframes: one for Personal Auto and one for Corporate Auto
personal_auto_df = df.loc[df['Policy Type'] == 'Personal Auto']
corporate_auto_df = df.loc[df['Policy Type'] == 'Corporate Auto']

print(f"Personal Auto customers: {len(personal_auto_df)}")
print(f"Corporate Auto customers: {len(corporate_auto_df)}")

# Calculate the average income for each policy type
personal_auto_avg_income = personal_auto_df['Income'].mean()
corporate_auto_avg_income = corporate_auto_df['Income'].mean()

# Print the results
print(f"\nAverage Income Comparison:")
print(f"Personal Auto average income: ${personal_auto_avg_income:,.2f}")
print(f"Corporate Auto average income: ${corporate_auto_avg_income:,.2f}")

# Calculate the difference
income_difference = corporate_auto_avg_income - personal_auto_avg_income
print(f"\nDifference: ${income_difference:,.2f}")

if personal_auto_avg_income < corporate_auto_avg_income:
    print("Result: Personal Auto customers have LOWER average income than Corporate Auto customers")
else:
    print("Result: Personal Auto customers do NOT have lower average income than Corporate Auto customers")
    
print(f"Corporate Auto customers earn {income_difference/personal_auto_avg_income*100:.1f}% more on average")
```

Consider customers with a high policy claim amount to be those in the top 25% of the total claim amount. Create a pandas DataFrame object that contains information about customers with a policy claim amount greater than the 75th percentile.
- Use DataFrame methods to calculate summary statistics about the high policy claim amount data. 

*Note: When analyzing data, we often want to focus on certain groups of values to gain insights. Percentiles are a useful tool to help us define these groups. A percentile is a measure that tells us what percentage of values in a dataset are below a certain value. For example, the 75th percentile represents the value below which 75% of the data falls. Similarly, the 25th percentile represents the value below which 25% of the data falls. When we talk about the top 25%, we are referring to the values that fall above the 75th percentile, which represent the top quarter of the data. On the other hand, when we talk about the bottom 25%, we are referring to the values that fall below the 25th percentile, which represent the bottom quarter of the data. By focusing on these groups, we can identify patterns and trends that may be useful for making decisions and taking action.*

*Hint: look for a method that gives you the percentile or quantile 0.75 and 0.25 for a Pandas Series.*

*Hint 2: check `Boolean selection according to the values of a single column` in https://towardsdatascience.com/filtering-data-frames-in-pandas-b570b1f834b9*

```python
# Your code here
# First, let's check the actual column names in the dataset
print("=== DATASET COLUMN NAMES ===")
print("Columns in the dataset:")
print(df.columns.tolist())

# Review statistics for total claim amount
print("\n=== TOTAL CLAIM AMOUNT STATISTICS ===")
print(df['Total Claim Amount'].describe())

# Calculate the 75th percentile (top 25% threshold)
percentile_75 = df['Total Claim Amount'].quantile(0.75)
percentile_25 = df['Total Claim Amount'].quantile(0.25)

print(f"\n25th Percentile (bottom 25%): ${percentile_25:.2f}")
print(f"75th Percentile (top 25% threshold): ${percentile_75:.2f}")

# Create DataFrame with customers having policy claim amount > 75th percentile
high_claim_customers = df[df['Total Claim Amount'] > percentile_75]

print(f"\nNumber of customers with high policy claim amounts (top 25%): {len(high_claim_customers)}")
print(f"Percentage of total customers: {len(high_claim_customers)/len(df)*100:.1f}%")

# Calculate summary statistics for high policy claim amount data
print("\n=== HIGH POLICY CLAIM AMOUNT CUSTOMERS - SUMMARY STATISTICS ===")

print("\nNumerical columns summary:")
numerical_cols = high_claim_customers.select_dtypes(include=[np.number]).columns
print(high_claim_customers[numerical_cols].describe())

print("\nCategorical columns summary:")
categorical_cols = high_claim_customers.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\n{col}:")
    print(high_claim_customers[col].value_counts())

print("\n=== INSIGHTS ON HIGH CLAIM CUSTOMERS ===")
print("Comparison with overall population:")

# Compare average values (only for numerical columns)
print(f"\nAverage Income:")
print(f"  High claim customers: ${high_claim_customers['Income'].mean():.2f}")
print(f"  All customers: ${df['Income'].mean():.2f}")

print(f"\nAverage Monthly Premium Auto:")
print(f"  High claim customers: ${high_claim_customers['Monthly Premium Auto'].mean():.2f}")
print(f"  All customers: ${df['Monthly Premium Auto'].mean():.2f}")

# Most common characteristics among high claim customers
print(f"\nMost common characteristics among high claim customers:")
print(f"  Most common state: {high_claim_customers['ST'].mode()[0]}")
print(f"  Most common policy type: {high_claim_customers['Policy Type'].mode()[0]}")
print(f"  Most common vehicle class: {high_claim_customers['Vehicle Class'].mode()[0]}")

# Check if Gender column exists with different name
if 'Gender' in df.columns:
    print(f"  Most common gender: {high_claim_customers['Gender'].mode()[0]}")
elif any('gender' in col.lower() for col in df.columns):
    gender_col = [col for col in df.columns if 'gender' in col.lower()][0]
    print(f"  Most common {gender_col}: {high_claim_customers[gender_col].mode()[0]}")

# Display first few rows of high claim customers (selected columns)
print(f"\n=== SAMPLE OF HIGH CLAIM CUSTOMERS ===")
key_cols = ['Customer', 'ST', 'Income', 'Total Claim Amount', 'Policy Type', 'Vehicle Class']
# Only include columns that exist
display_cols = [col for col in key_cols if col in df.columns]
print(high_claim_customers[display_cols].head())

print(f"\n=== KEY FINDINGS ===")
print(f" {len(high_claim_customers)} customers ({len(high_claim_customers)/len(df)*100:.1f}%) have high claim amounts (>${percentile_75:.2f})")
print(f" High claim customers have {'higher' if high_claim_customers['Income'].mean() > df['Income'].mean() else 'lower'} average income than overall population")
print(f" High claim customers pay {'higher' if high_claim_customers['Monthly Premium Auto'].mean() > df['Monthly Premium Auto'].mean() else 'lower'} monthly premiums on average")
print(f" Most high claim customers are from: {high_claim_customers['ST'].mode()[0]}")
print(f" Most common policy type among high claimers: {high_claim_customers['Policy Type'].mode()[0]}")
print(f" Most common vehicle class among high claimers: {high_claim_customers['Vehicle Class'].mode()[0]}")
```
